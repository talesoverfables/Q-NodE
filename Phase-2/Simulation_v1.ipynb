# Visualizing Bloch Sphere
Visualizing the state of a single Qubit




```python
!pip install qutip
```

```python
import numpy as np
import qutip as qt
```

```python
# Create a Bloch sphere instance
b = qt.Bloch()
```

```python
# Add a single data point (vector)
vec = [1, 0, 0]
b.add_vectors(vec)
```

```python
# Add a quantum state (e.g., |up> state)
up = qt.basis(2, 0)
b.add_states(up)
```

```python
# Add a single data point (coordinates)
pnt = [1 / np.sqrt(3), 1 / np.sqrt(3), 1 / np.sqrt(3)]
b.add_points(pnt)
```

```python
# Render the Bloch sphere
b.show()
```

#  Bloch Sphere Representations of Q Gates

## Pauli-X (NOT Gate)

```python
# Define a single qubit state |0⟩
initial_state = qt.basis(2, 0)
```

```python
# Apply the Pauli-X gate (NOT gate)
x_gate = qt.sigmax()
final_state = x_gate * initial_state
```

```python
# Print the results
print("Initial state (|0⟩):", initial_state)
print("Final state after applying X gate:", final_state)
```

```python
# Create a Bloch sphere
bloch = qt.Bloch()

# Add the initial state and the final state to the Bloch sphere
bloch.add_states(initial_state)
bloch.add_states(final_state)
```

```python
# Render the Bloch sphere
bloch.show()
```

## Pauli-Y Gate


```python
# Create a Bloch sphere
bloch1 = qt.Bloch()
```

```python
# Define the initial state |0⟩
initial_state1 = qt.basis(2, 0)  # You can also use qt.basis(2, 1) for |1⟩
```

```python
print(initial_state1)
```

```python
# Apply the Pauli-Y gate
y_gate = qt.sigmay()
final_state1 = y_gate * initial_state1
print(final_state1)
```

```python
# Add the initial state and the final state to the Bloch sphere
bloch1.add_states(initial_state1, kind='vector')
bloch1.add_states(final_state1, kind='vector')
```

```python
# Render the Bloch sphere
bloch1.show()
```

## Hadamard (H) Gate



```python
# Create a Bloch sphere
bloch2 = qt.Bloch()
```

```python
# Define the initial state |0⟩
initial_state2 = qt.basis(2, 0)  # You can also use qt.basis(2, 1) for |1⟩
```

```python
# Apply the Hadamard gate
h_gate = 1 / np.sqrt(2) * (qt.qeye(2) + qt.sigmax())
final_state2 = h_gate * initial_state2
print(final_state2)
```

```python
# Add the initial state and the final state to the Bloch sphere
bloch2.add_states(initial_state2, kind='vector')
bloch2.add_states(final_state2, kind='vector')
```

```python
# Render the Bloch sphere
bloch2.show()
```

## Pauli-Z Gate

```python
# Create a Bloch sphere
bloch3 = qt.Bloch()
```

```python
# Define the initial state |1⟩
initial_state3 = qt.basis(2, 1)
```

```python
# Add the initial state to the Bloch sphere
bloch3.add_states(initial_state3)
```

```python
# Add the Pauli Z gate effect (rotation around Z-axis by π radians)
pauli_z_gate = qt.qeye(2) - 2 * qt.basis(2, 1) * qt.basis(2, 1).dag()
final_state3 = pauli_z_gate * initial_state3
```

```python
# Render the Bloch sphere
bloch3.show()
```
